<h3 id="2.1-|-Load-Modules"><b><span id="subtitle">2.1</span> | Load Modules</b></h3>

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
from matplotlib import colors
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import seaborn as sns
import sqlite3

import os
```

```python
#define working directory for use in path to the data file
working_dir = os.getcwd()
#load the dataset into a pandas dataframe through sqlite3
path = working_dir + '/data/FPA_FOD_20210617.sqlite'
con = sqlite3.connect(path)
fires = pd.read_sql_query("SELECT * FROM fires", con, parse_dates = ['DISCOVERY_DATE','CONT_DATE'])
```

```python
#examine data
fires.head()
```

<div class="description">
    <h3 id="2.2-|-Drop-Columns"><b><span id="subtitle">2.2</span> | Drop Columns</b></h3>
     <p>Many columns are irrelevant or redundant, and would introduce unnecessary noise, such as "source_reporting_unit_name" and "source_reporting_unit", or "ics_209_plus_complex_join_id", which is just an ID to keep track of fires in the forest service database. Many columns are not clear based on their name alone, so reviewing the Forest Service's description was useful in determining what to drop.
    </p>
</div>


```python
# We are only interested in 10 columns, we will drop the rest from the dataset
# We will also convert the column names to lowercase
final_cols = ['FIRE_NAME', 'DISCOVERY_DATE', 'NWCG_GENERAL_CAUSE', 'CONT_DATE', 'FIRE_SIZE', 
             'OWNER_DESCR', 'STATE', 'FIRE_SIZE_CLASS', 'LATITUDE','LONGITUDE']
fires.drop(columns = fires.columns.difference(final_cols), inplace=True)
fires.columns = fires.columns.str.lower()
```

```python
#print the data types and then the shape of our data
#want to ensure the dates were parsed correctly
print(fires.dtypes)
print (f'\nSHAPE: {fires.shape}')
```

```python
#check which columns contain null values
fires.isna().sum()
```

```python
#fill N/A values for fire_name 
fires['fire_name'].fillna('Unkown', inplace = True)
```

```python
#we will replace the N/A containment dates with the discovery dates as this information isn't important here
fires.cont_date.fillna(fires.discovery_date, inplace = True)
```

```python
#describe our dataframe, since most info is categorical we will only see fire size and fire coordinates here
fires.describe()
```

<h1 id="3.-Visualization-and-Analaysis"><span class="section_num">3.</span> Visualization and Analaysis</h1>

<div class="description">  
   <h3 id="3.1-|-Total-Number-of-Fires-and-Total-Fire-Acreage"><b><span id="subtitle">3.1</span> | Total Number of Fires and Total Fire Acreage</b></h3>
        <p> These pie charts represent states with the largest number of individual fires, and the states with the most acreage burnt. Many states have a large number of fires, but not a large amount of acreage burnt, or vice-vera, which we will explore further.
        </p>
</div>

```python
#create dataframe to hold the states with the highest number of fires
most_fires = fires.state.value_counts().sort_values(ascending = False).nlargest(10).to_frame()
#create dataframe to hold the states with the most acreage burnt
most_acreage = fires.groupby('state').fire_size.sum().sort_values(ascending = False).nlargest(10).to_frame()

#define sns color palettes for each pie chart
color1 = sns.diverging_palette(10, 20, center = "light", s = 75, l = 50)
color2 = sns.diverging_palette(145, 37,center="light", s=50, l = 60)

fig, axes = plt.subplots(1,2, figsize=(30, 30))
#pie chart for the states with the most number of fires
axes[0].pie(most_fires.state, labels = most_fires.index, colors = color1,
            autopct = lambda x: np.round(x/100.*most_fires.state.sum(), 0),
            textprops={'fontsize': 20})
#pie chart for the states with the most acreage burnt
axes[1].pie(most_acreage.fire_size, labels = most_acreage.index, colors = color2,
            autopct = lambda x: np.round(x/100.*most_acreage.fire_size.sum(), 0),
            textprops={'fontsize': 20})

axes[0].set_title('Number of Fires per State',fontweight = 'bold', size = 20)
axes[0].text(-0.30,0.90, 'Number of Fires', ha="right", va="top", transform=plt.gca().transAxes)

axes[1].set_title('Total Fire Acreage per State',fontweight = 'bold', size = 20)
axes[1].text(0.95,0.92, 'Acreage Burnt', ha="right", va="top", transform=plt.gca().transAxes)

plt.show()
```

<div class="analysis">
    <p>The fact that many states show up in only one of the two groups tells us that having a large number of fires doesn't necesarily equate to a large amount of acreage burnt, or vice-versa. For example Alaska makes up almost a quarter of the total acreage burnt in the country, but it isn't even in the top ten in terms of number of fires. I would imagine that this is probably because Alaska has a much larger percentage of wild and unmanaged forests compared to the rest of the country and we will explore this theory with our next chart.
    </p>
</div>

<hr>

<div class="description">    
    <h3 id="3.2-|-Fire-Cause-in-Alaska-and-Georgia-versus-Other-States"><b><span id="subtitle">3.2</span> | Fire Cause in Alaska and Georgia versus Other States</b></h3>
        <p>Next we will create a barplot of fire causes in Alaska and Georgia compared to the rest of the US to get an idea of how states with a large amount of acreage burnt compare to states with a large number of fires. Alaska and Georgia are chosen because they are the largest out of states that are in only one of the two groups.
        </p>
</div>

```python
#get normalized value counts of fire causes in Alaska
ak_causes = fires[fires['state'] == "AK"].nwcg_general_cause.value_counts(normalize = True)
#get normalized value counts of fire causes in Georgia
ga_causes = fires[fires['state'] == "GA"].nwcg_general_cause.value_counts(normalize = True)

#get normalized value counts of fire causes in the rest of the US
contiguous_causes = fires.nwcg_general_cause.value_counts(normalize = True)
#create a new DF with the percent causes for Alaska and percent for other states as columns
cause_df = pd.concat([ak_causes.rename('AK_percent'), ga_causes.rename('GA_percent'),
                contiguous_causes.rename('contiguous_US_percent')], axis=1)

fig, ax = plt.subplots(figsize=(24,10))
#create a barplot for AK, GA, and Other States on the same axis to create a stacked barplot
ax = sns.barplot(x=cause_df.index, y=cause_df["AK_percent"], color='dodgerblue', alpha = .7, label = "Alaska")
ax = sns.barplot(x=cause_df.index, y=cause_df["GA_percent"], color='red', alpha = .7, label = "Georgia")
ax = sns.barplot(x=cause_df.index, y=cause_df["contiguous_US_percent"], color='gray', alpha = .7, label = "Other States")

ax.set_title('Cause of Fires in Alaska versus Other States',fontsize=20)
ax.tick_params(axis='x', labelrotation= 40, labelsize=15)
ax.set_ylabel("Percentage per Fire Cause")
ax.legend(prop={'size': 20})
```

<div class="analysis">
    <p>From the barplot we can confirm that Alaska has a much higher percentage of fires attributed to natural causes, which supports the hypothesis that Alaskan fires are much larger because of its vast amounts of untamed wilderness. We also see that Georgia has a much higher percentage of fires, over fifty percent, attributed to debris and open burning, and not even ten percent attribution to natural causes. This tells us that a large amount of small fires are caused by debris and open burning, or human impact, whereas large fires are most commonly started naturally.
    </p>
</div>

<div class="analysis">    
        <p>If we look at the percentage of different causes for the largest 1,000 fires in US history we can see that 70% were attributed to natural causes, which further reinforces the theory that large fires are most likely to be caused naturally.
        </p>
</div>

```python
#group the 1000 largest fires together
largest_fires = fires.sort_values('fire_size', ascending = False).head(1000)
#calculate the percentage of each cause by normalizing the value counts
pct_causes = largest_fires['nwcg_general_cause'].value_counts(normalize=True) * 100
print(pct_causes)
```

<div class="analysis">    
        <p>Now if we look at the 100,000 smallest fire we see that natural causes make up only 7.6% percent of small fires with roughly 50% attributed to some kind of human impact. 
        </p>
</div>

```python
#group the 100000 smallest fires together
smallest_fires = fires.sort_values('fire_size', ascending = True).head(100000)
#calculate the percentage of each cause by normalizing the value counts
pct_causes_small = smallest_fires['nwcg_general_cause'].value_counts(normalize=True) * 100
print(pct_causes_small)
```

<div class="description">    
    <h3 id="3.4-|-Fire-Size-Map"><b><span id="subtitle">3.4</span> | Fire Size Map</b></h3>
        <p>By creating a scatter plot where each point corresponds to a fire's latitude and longitude and the size and color of the point represents the fire's size, it is relatively simple to visualize how fire size changes across the country.
            </p>
</div>

```python
#to make our map more legible we will constrain the latitude and longitude of our fire data to ignore outlying areas and wasted space

#this function is used to move Alaskan fire's longitude to the right for easier visualization
def transpose_AK(longitude):
    if longitude < -130:
        longitude += 20
    return longitude

cropped_coordinates = fires.copy()
#apply our tranpose function to the latitude and longitude of ALaska
cropped_coordinates['longitude'] = cropped_coordinates['longitude'].apply(lambda x: transpose_AK(x))
#we further reduce the latitude and longitude of the dataframe in order to igonore Hawaii and other outlying areas
cropped_coordinates = cropped_coordinates[(cropped_coordinates['longitude'] > -145)
                          & (cropped_coordinates['latitude'] > 25)]

#create map
fig, ax = plt.subplots(figsize=(30, 20))
scatter = ax.scatter(data = cropped_coordinates, x="longitude", y="latitude", alpha=.8,cmap = "jet",
          c = cropped_coordinates["fire_size"] * 10,marker = 'o', s = cropped_coordinates["fire_size"] / 200,
          label = None, norm = colors.Normalize(vmin=cropped_coordinates['fire_size'].min(),
                           vmax=cropped_coordinates['fire_size'].max())
                    )
ax.set_title('Fires Size Across the US',{'fontsize':20})
```

<div class="analysis">
    <b>The map provides some interesting insight:</b>
    <ul>  
      <li style="font-size:17px"> It confirms our findings earlier about where the largest fires are concentrated. </li>
      <li style="font-size:17px"> It shows that the East Coast has almost no large fires at all, but a large amount of small fires,<br> which we can see from the density of dark blue points. </li>
      <li style="font-size:17px"> We can also see that the  Midwest has the least amount of fires in the country. </li>
      <li style="font-size:17px"> Areas closer to the Oregon and Washington coast also have a relatively small amount of fires,<br> especially when compared with
            other areas that are just a few hundred miles East or South. </li>
    </ul>
</div>

<div class="description">    
    <h3 id="3.5-|-Evolution-of-Fires-Over-Time"><b><span id="subtitle">3.5</span> | Evolution of Fires Over Time</b></h3>
        <p>Next I wanted to gauge how fires have changed over time. First we will look at a barplot of total acreage burnt per year, then we will examine how fire size changes each month using a heatmap and a boxplot.
        </p>
</div>

```python
#group fires together by year
fires_by_year = fires.copy()
#assign a year column based on the discovery_date datetime object
#the datetime object makes it simple to access the month or year from a date using a list comprehension
fires_by_year['year'] = [i.year for i in fires_by_year.discovery_date]
fires_by_year = fires_by_year.groupby('year').sum().reset_index()

sns.set(font_scale=1.5)
fig, ax = plt.subplots(figsize=(20, 12))
ax = sns.barplot(data = fires_by_year, x = 'year', y = "fire_size")
ax.set_xlabel('')
ax.set_ylabel('Acreage Burnt', fontsize = 20)
```

<div class="analysis">
Based on the barplot we can see that the amount of acreage burnt per year has a slight upward trend and the distribution is skewed slightly to the left. From this we can conclude that fires are getting worse as time moves forward.
</div>

<hr>

<div class="description">    
    <h3 id="3.6-|-Evolution-of-Fire-Acreage-by-Month"><b><span id="subtitle">3.6</span> | Evolution of Fire Acreage by Month</b></h3>
        <p>Next we will examine which months and years have the highest concentration of fires. We will also look at how fire density in each month has changed over time.<br><br>
            I was curious if fires are happening later in the year now than they used to, as I have noticed the fire season seems to be shifting later and later where I live in California, although this has been especially apparent the last two years and this dataset ends in 2018 unfortunately.
        </p>
</div>

```python
#we will create a new dataframe that will track the year and month of each fire
month_df = fires.copy()
month_df['month'] = [i.month for i in month_df.discovery_date]
month_df['year'] = [i.year for i in month_df.discovery_date]

# get the sum of fires per month
month_df = month_df.groupby(['month', 'year']).fire_size.sum()
month_df = month_df.unstack(level = 0)

# create heatmap
fig, ax = plt.subplots(figsize=(20, 15))
sns.heatmap(month_df, cmap = "Reds", linewidth=0.3, cbar_kws={"shrink": .7, 'label' : "Fire Acreage per Month"})
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(12) + .5, labels = xticks_labels)
plt.xlabel('')
plt.ylabel('')
ax.set_title('Total Fire Acreage per Month over Time',{'fontsize':20})
plt.show()
```

<div class="analysis">
    <p>
        The heatmap gives us a great representation of how fires are distributed throughout the season over each year in our dataset. 
<br><br>
            We can see that fires are the worst in June, July, and August, which is what we would expect.We can also see that a few years had larger fire activity in March and April and that the whole calendar year had more fire activity the last few years. 
<br><br>
            We can also see that the amount of acres being burnt per month is much worse in the past 15 years than the first 10 years of our dataset, which reinforces our earlier assessment that fires are happening on a larger scale each year.
    </p>
</div>

<hr>

<div class="description">    
    <h3 id="3.7-|-Fire-Density-by-Month"><b><span id="subtitle">3.7</span> | Fire Density by Month</b></h3>
        <p>Now we will use a box and strip plot to look at the distribution of fires per month, where each point from the strip plot represents the total acreage burnt in one month for each year in our dataset. 
        </p>
</div>

```python
#using month_df from heatmap for our data here
fig, ax = plt.subplots(figsize=(10, 15))
ax = sns.boxplot(data = month_df)
ax = sns.stripplot(data=month_df, color=".25", s = 5)
xticks_labels = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
_ = plt.xticks(np.arange(12), labels = xticks_labels)
plt.ylabel('Total Acreage per Month')
```

<div class="analysis">
    <p> We can clearly see that June, July, and August are the months with the most acreage burnt, which was reflected in our heatmap as well. However we can get a better idea of the distribution of fires here and gain some valuable insight:
    </p>
    <ul>
        <li> First of all we can see that acreage burnt is clustered tightly together for most of spring and winter, and fires gradually ramp up from March to May before doubling in size in June. </li>
        <li> June, July, and August all have roughly the same median, but July's third quartile is much larger than June or August's, which means we have greater dispersion of large fires in July.</li>
        <li> June has the largest outliers which represent exceptionally large amounts of acreage burnt in a month.</li>
    </ul>
</div>
